```python
def filter_df(df):
    keep_columns = [
        "exp_id",
        "agent_id",
        "num_episodes",
        "num_sims",
        "policy_name",
        "nesting_level",
        "episode_discounted_returns_mean",
        "episode_discounted_returns_CI",
    ]
    f_df = df[keep_columns]
    f_df = f_df.sort_values(by=["exp_id", "agent_id"])
    f_df.rename(
        columns={
            "episode_discounted_returns_mean": "Mean Disc. Return",
            "episode_discounted_returns_CI": "Mean Disc. Return CI"
        },
        inplace=True
    )
    
    f_df["Time (s)"] = df["search_time_mean"] + df["update_time_mean"]
    f_df["Time (s) std"] = (df["search_time_std"]**2 + df["update_time_std"]**2)**1/2
    f_df["Time (s) CI"] = 1.96 * (f_df["Time (s) std"] / (df["num_episodes"])**1/2)
    
    return f_df
```
